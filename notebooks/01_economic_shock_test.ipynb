# AI Civilization Builder: Economic Shock Test

This notebook demonstrates how to run an automated simulation experiment to study the effects of a food supply drought (economic shock) on market clearing prices and citizen survival rates.

```python
import sys
import os
# Add project root to path
sys.path.append(os.path.abspath('../'))

from backend.app.core.engine import CivilizationEngine
from backend.app.simulation.economy import MarketBook, Order
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Setup Simulation World

```python
engine = CivilizationEngine(
    world_id="experiment-shock-uuid",
    name="Research Colony",
    seed=42,
    width=20,
    height=20
)

# Spawn 10 citizens
for i in range(10):
    engine.add_citizen(f"Researcher {i}", age=20+i, wealth=100.0)
```

## 2. Baseline Period (Ticks 1-15)
Run the simulation under normal conditions where food resources naturally exist on plains.

```python
baseline_prices = []
for tick in range(15):
    engine.step()
    # Sample current food clearing price
    baseline_prices.append(engine.spatial_map.food_nodes.mean())

print(f"Baseline average food density: {sum(baseline_prices)/len(baseline_prices):.3f}")
```

## 3. Introduce Economic Shock: Severe Drought (Tick 16)
Set all natural food resources on the map to zero.

```python
print("Applying economic shock: clear all food nodes!")
engine.spatial_map.food_nodes.fill(0.0)

shock_prices = []
for tick in range(15):
    engine.step()
    # Price will rise as agents compete for remaining food inventory
    shock_prices.append(engine.spatial_map.food_nodes.mean())

print("Simulation completed.")
```
